```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

Exercise 1

```python
#import matplotlib
import matplotlib.pyplot as plt
#Create the figure and plot the graph
plt.figure()
plt.plot(x,y)
#Assign lables to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
```

```python
#Create figure object
fig=plt.figure()
#Plot the axes ax1 and ax2
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
```

```python
#Create figure object
fig=plt.figure()
#Plot the axes ax1 and ax2
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
#Display plot for each axes
ax1.plot(x,y)
ax2.plot(x,y)
```

Exercise 3

```python
#Create figure object
fig=plt.figure()
#Plot the axes ax1 and ax2
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
ax2.set_xlim(20.0,22.0)
ax2.set_ylim(30,50)
#Display plot for each axes
ax1.plot(x,z,)
ax2.plot(x,y)
#Set labeles
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
```

```python
#PLot in subplots
fig, ax = plt.subplots(1, 2)
ax[0].plot(x,y,color='b',ls='--')
ax[1].plot(x,z,color='r')
```
